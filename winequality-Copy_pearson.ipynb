importing the sklearn library to perform the feature selection operation

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import f_regression,SelectKBest
```

```python
x=df.iloc[:,:-1]
y=df['quality']
x
```

performing the feature selection using the f_regression with select k best method
visit the website for more info 
f_regression- https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.f_regression.html
selectKBest - https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectKBest.html

```python
def feature_selection(x_train,x_test,y_train):
    fs=SelectKBest(score_func=f_regression,k=4)
    fs.fit(x_train,y_train)
    
    return fs
```

selecting the top 4 columns which shows more impact on the quality of wine according to their scores  

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
fs=feature_selection(x_train,x_test,y_train)
for i in range(len(fs.scores_)):
    print('Feature %d: %f %s' % (i, fs.scores_[i],df.columns[i]))
```

removing the outliers in the dataset with percentile method

```python
Q1 = new_df.iloc[:,:-1].quantile(0.25)
Q3 = new_df.iloc[:,:-1].quantile(0.75)
IQR = Q3 - Q1

new_df = new_df[~((new_df.iloc[:,:-1] < (Q1 - 1.5 * IQR)) |(new_df.iloc[:,:-1] > (Q3 + 1.5 * IQR))).any(axis=1)]
```
